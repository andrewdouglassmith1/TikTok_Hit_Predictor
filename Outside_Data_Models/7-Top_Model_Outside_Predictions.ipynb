```python
import pickle
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import os
from sklearn.preprocessing import StandardScaler
```

# Pickle in Data

```python
# Pickle in cleaned dataframe

# Designate path

path = r"C:\Users\Andrew\Documents\Metis\TikTok_Song_Predictor\Pickle\Kaggle.pkl"

df = pickle.load(open(path,'rb'))
df.head(2)
```

# KNN

```python
df.columns
```

```python
# Seperate features from label

X = df.loc[:,['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','year']]
```

```python
# Pickle in model

path = r"C:\Users\Andrew\Documents\Metis\TikTok_Song_Predictor\Pickle\knn_model.pkl"

model = pickle.load(open(path,'rb'))
```

```python
# Scale features 
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Predict outcomes
y_predict = model.predict(X)
```

```python
# Predict proba
y_predict_proba = model.predict_proba(X)
```

```python
# Send predictions to dataframe

df_predict = pd.DataFrame(y_predict, columns = ['success'] )
```

```python
# Send predict proba to dataframe

df_predict_proba = pd.DataFrame(y_predict_proba, columns = ['prob not a success', 'prob a success'])
```

```python
# Merge both dataframes

df_agg_knn = pd.concat([df, df_predict_proba], axis =1)
```

```python
df_agg_knn = pd.concat([df_agg_knn, df_predict], axis=1)
```

```python
df_counts = df_agg_knn['success']
```

```python
df_counts.value_counts()
```

```python
# Sort by prob a success

df_agg_knn = df_agg_knn.sort_values(by=['prob a success'],ascending =  False, ignore_index = True)
```

```python
# Create a column that is the spotify link 

def spotify_link(song_id):
    return f"https://open.spotify.com/embed/track/{song_id}"
```

```python
df_agg_knn['link'] = df_agg_knn['id'].apply(lambda x: spotify_link(x))
```

# Pickle Dataframe

```python
# Designate path

path = r"C:\Users\Andrew\Documents\Metis\TikTok_Song_Predictor\Pickle\Kaggle_Results_knn.pkl"
with open(path, 'wb') as f:
    pickle.dump(df_agg_knn, f)
```

```python
# Save to csv
path = r'C:\\Users\\Andrew\\Documents\\Metis\\TikTok_Song_Predictor\\Streamlit\\'
```
